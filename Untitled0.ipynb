```python
import os
import cv2

input_folder = '/content/iris/'
output_folder = '/content/processed_results/'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

valid_extensions = ('.jpg', '.jpeg', '.png', '.bmp')

if not os.path.exists(input_folder):
    print(f"Ошибка: Папка {input_folder} не найдена!")
else:
    all_files_paths = []
    for root, dirs, files in os.walk(input_folder):
        for file in files:
            if file.lower().endswith(valid_extensions):
                all_files_paths.append(os.path.join(root, file))

    total_files = len(all_files_paths)
    print(f"Найдено файлов: {total_files}")

    success_count = 0
    error_count = 0

    for index, input_path in enumerate(all_files_paths):
        filename = os.path.basename(input_path)
        output_path = os.path.join(output_folder, f'processed_{index}_{filename}')
        try:
            enhanced_img, circles = preprocess_iris(input_path)
            cv2.imwrite(output_path, enhanced_img)
            success_count += 1
        except Exception as e:
            error_count += 1

    print(f"Обработка завершена. Успешно: {success_count}, Ошибок: {error_count}")
```

```python
from skimage.feature import local_binary_pattern
import numpy as np

def extract_lbp_features(gray_image, P=8, R=1):
    lbp = local_binary_pattern(gray_image, P, R, method="uniform")
    n_bins = int(lbp.max() + 1)
    hist, _ = np.histogram(lbp.ravel(), bins=n_bins, range=(0, n_bins), density=True)
    return hist
```

```python
import torch
import torch.nn as nn

class IrisNet(nn.Module):
    def __init__(self, num_classes):
      super(IrisNet, self).__init__()
      self.features = nn.Sequential(
          nn.Conv2d(1, 32, kernel_size=3, padding=1),
          nn.ReLU(),
          nn.MaxPool2d(2, 2),
          nn.Conv2d(32, 64, kernel_size=3, padding=1),
          nn.ReLU(),
          nn.MaxPool2d(2, 2)
      )
      self.classifier = nn.Sequential(
          nn.Flatten(),
          nn.Linear(64 * (image_size // 4) * (image_size // 4), 128),
          nn.ReLU(),
          nn.Dropout(0.5),
          nn.Linear(128, num_classes)
      )

    def forward(self, x):
      x = self.features(x)
      x = self.classifier(x)
      return x
```

```python
from sklearn.ensemble import StackingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

# Базовые классификаторы
estimators = [
    ('rf', RandomForestClassifier(n_estimators=50)),
    ('svc', SVC(probability=True))
]

# Мета-классификатор (логистическая регрессия объединяет ответы)
stacking_model = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression()
)

# Как использовать:
# stacking_model.fit(X_train, y_train)
# predictions = stacking_model.predict(X_test)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from sklearn.preprocessing import LabelEncoder
import cv2
import os
import numpy as np

# Восстановим функцию предобработки, так как она необходима для пайплайна
def preprocess_iris(image_path, size=(64, 64)):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None: return None, None
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    enhanced_img = clahe.apply(img)
    enhanced_img = cv2.resize(enhanced_img, size)
    return enhanced_img, None

class IrisDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.image_paths = []
        self.labels = []
        self.classes = []

        # Собираем пути и метки (имя папки = имя класса)
        for class_name in sorted(os.listdir(root_dir)):
            class_path = os.path.join(root_dir, class_name)
            if os.path.isdir(class_path):
                self.classes.append(class_name)
                for img_name in os.listdir(class_path):
                    if img_name.lower().endswith(('.jpg', '.png', '.jpeg')):
                        self.image_paths.append(os.path.join(class_path, img_name))
                        self.labels.append(class_name)

        self.label_encoder = LabelEncoder()
        self.encoded_labels = self.label_encoder.fit_transform(self.labels)

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        label = self.encoded_labels[idx]

        # 1. Pipeline предобработки
        img, _ = preprocess_iris(img_path)
        if img is None: # Заглушка на случай ошибки чтения
            img = np.zeros((64, 64), dtype=np.uint8)

        # 2. Извлечение LBP признаков для классики
        lbp_features = extract_lbp_features(img)

        # 3. Подготовка для нейросети
        if self.transform:
            img_tensor = self.transform(img)
        else:
            img_tensor = torch.from_numpy(img).float().unsqueeze(0) / 255.0

        return img_tensor, torch.tensor(lbp_features, dtype=torch.float32), torch.tensor(label, dtype=torch.long)
```

```python
# Инициализация данных
image_size = 64 # Определяем размер для IrisNet
dataset = IrisDataset(root_dir='/content/iris/iris/')
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

num_classes = len(dataset.classes)
model_cnn = IrisNet(num_classes=num_classes)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model_cnn.parameters(), lr=0.001)

print(f"Загружено классов: {num_classes}")
print(f"Всего изображений: {len(dataset)}")

# Краткий цикл обучения CNN
model_cnn.train()
for epoch in range(5):
    for imgs, lbps, labels in dataloader:
        optimizer.zero_grad()
        outputs = model_cnn(imgs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
    print(f"Эпоха {epoch+1}, Loss: {loss.item():.4f}")
```

```python
# Обучение StackingClassifier на комбинации признаков
model_cnn.eval()
all_cnn_preds = []
all_lbp_features = []
all_labels = []

with torch.no_grad():
    for imgs, lbps, labels in dataloader:
        outputs = model_cnn(imgs)
        # Берем вероятности из CNN как дополнительные признаки
        probs = torch.softmax(outputs, dim=1).numpy()
        all_cnn_preds.append(probs)
        all_lbp_features.append(lbps.numpy())
        all_labels.append(labels.numpy())

X_stack = np.hstack([np.vstack(all_lbp_features), np.vstack(all_cnn_preds)])
y_stack = np.concatenate(all_labels)

print("Обучение Stacking модели на комбинированных признаках...")
stacking_model.fit(X_stack, y_stack)
print("Система готова!")
```

```python
def identify_iris(image_path, model_cnn, stacking_model, dataset):
    img, _ = preprocess_iris(image_path)
    if img is None: return "Ошибка загрузки"

    lbp_feat = extract_lbp_features(img).reshape(1, -1)

    model_cnn.eval()
    with torch.no_grad():
        img_tensor = torch.from_numpy(img).float().unsqueeze(0).unsqueeze(0) / 255.0
        cnn_output = model_cnn(img_tensor)
        cnn_probs = torch.softmax(cnn_output, dim=1).numpy()

    X_final = np.hstack([lbp_feat, cnn_probs])
    class_idx = stacking_model.predict(X_final)[0]
    return dataset.label_encoder.inverse_transform([class_idx])[0]
```

```python
from sklearn.metrics import accuracy_score

# Оценка на всем текущем наборе (X_stack, y_stack уже созданы в предыдущих шагах)
y_pred = stacking_model.predict(X_stack)
accuracy = accuracy_score(y_stack, y_pred)

print(f"Точность (Accuracy) на текущем наборе данных: {accuracy*100:.2f}%")
```
